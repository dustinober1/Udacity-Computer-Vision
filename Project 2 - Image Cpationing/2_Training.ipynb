### Question 1

**Question:** Describe your CNN-RNN architecture in detail.  With this architecture in mind, how did you select the values of the variables in Task 1?  If you consulted a research paper detailing a successful implementation of an image captioning model, please provide the reference.

**Answer:** 

The CNN-RNN architecture consists of two main components: the CNN-based encoder and the LSTM-based decoder.

The encoder is implemented as an instance of the EncoderCNN class. It uses a pre-trained ResNet-50 model as the backbone to extract features from input images. The last fully connected layer of ResNet-50 is removed and replaced with a linear layer to reduce the output size to embed_size. The final feature representation is obtained by applying the linear layer to the flattened features tensor. The forward method takes in the images tensor as input and returns the extracted features tensor.

The decoder is implemented as an instance of the DecoderRNN class. It takes in the extracted features tensor from the encoder and the captions tensor as input. The captions tensor is first embedded using an embedding layer that maps word indices to their embeddings. The embedded features tensor and the embedded captions tensor are then concatenated and fed into an LSTM that takes in the embedded word vectors as input and outputs hidden states at each time step. The final output scores are obtained by applying a linear layer to the hidden states. The forward method takes in the features and captions tensors as input and returns the output scores tensor.

The sample method is used to generate predicted sentences given a pre-processed input image tensor. It takes in the pre-processed input image tensor, initial hidden state of the LSTM, and the maximum length of the predicted sentence as input. The method uses the LSTM to generate the predicted word indices for the output sentence. It repeatedly feeds the predicted word embeddings back into the LSTM until the maximum length is reached or the EOS token is predicted.

The transform_train is used to apply data augmentation and preprocessing techniques on the training images before passing them to the model.

- RandomCrop: randomly crops the image to a specified size (224x224 pixels)
- RandomHorizontalFlip: flips the image horizontally with a probability of 0.5
- ToTensor: converts the image to a PyTorch tensor
- Normalize: normalizes the image with the mean and standard deviation of the ImageNet dataset

These transforms are commonly used in image classification tasks and are considered to be a good choice for training CNN models. The RandomCrop and RandomHorizontalFlip transforms help increase the variability of the training data and improve the model's ability to generalize to new images. The ToTensor transform converts the image to a tensor, which is the expected input format for the CNN model. Finally, the Normalize transform scales the pixel values to have a mean of [0.485, 0.456, 0.406] and a standard deviation of [0.229, 0.224, 0.225], which are the mean and standard deviation of the ImageNet dataset. This normalization step helps the model converge faster during training.

Overall, the transform_train provided in the code is a good choice for the CNN architecture because it applies a set of common image preprocessing and augmentation techniques that have been shown to be effective for training CNNs on image classification tasks.

### Question 3

**Question:** How did you select the trainable parameters of your architecture?  Why do you think this is a good choice?

**Answer:** 

In the encoder part of the architecture, I loaded a pre-trained ResNet-50 model and disabled learning for all parameters in the backbone by setting param.requires_grad_(False). This is a common practice in transfer learning where a pre-trained model is fine-tuned on a specific task. Since the ResNet-50 model was trained on ImageNet dataset, it has already learned to recognize and extract high-level features from images. By disabling the learning of these parameters, we ensure that the already learned features are preserved, and we only need to learn the new linear layer that reduces the output size to embed_size.

In the decoder part of the architecture, we have three trainable parameters: embed, lstm, and linear. The embed layer maps word indices to their embeddings, which are then passed to the lstm layer. The lstm layer takes in the embedded word vectors as input and outputs hidden states at each time step. Finally, the linear layer maps the hidden states to the vocabulary size.

The choice of these trainable parameters is based on the architecture of the LSTM-based image captioning model. The embed layer is used to learn the embeddings of the input captions, while the lstm layer is used to learn the temporal dependencies between the input embeddings and generate the output sequence. The linear layer is used to map the hidden states to the vocabulary size, allowing the model to generate a probability distribution over the vocabulary at each time step.

Overall, these trainable parameters are a good choice because they allow the model to learn the appropriate representations for the input images and captions, and to generate meaningful and accurate captions for the input images.

````python
import torch
import torch.nn as nn
from torchvision import transforms
import sys
sys.path.append('/opt/cocoapi/PythonAPI')
from pycocotools.coco import COCO
from data_loader import get_loader
from model import EncoderCNN, DecoderRNN
import math


## TODO #1: Select appropriate values for the Python variables below.
batch_size = 128          # batch size
vocab_threshold = 5        # minimum word count threshold
vocab_from_file = True    # if True, load existing vocab file
embed_size = 512           # dimensionality of image and word embeddings
hidden_size = 512          # number of features in hidden state of the RNN decoder
num_epochs = 3             # number of training epochs
save_every = 1             # determines frequency of saving model weights
print_every = 100          # determines window for printing average loss
log_file = 'training_log.txt'       # name of file with saved training loss and perplexity

# (Optional) TODO #2: Amend the image transform below.
transform_train = transforms.Compose([ 
    transforms.Resize(256),                          # smaller edge of image resized to 256
    transforms.RandomCrop(224),                      # get 224x224 crop from random location
    transforms.RandomHorizontalFlip(),               # horizontally flip image with probability=0.5
    transforms.ToTensor(),                           # convert the PIL Image to a tensor
    transforms.Normalize((0.485, 0.456, 0.406),      # normalize image for pre-trained model
                         (0.229, 0.224, 0.225))])

# Build data loader.
data_loader = get_loader(transform=transform_train,
                         mode='train',
                         batch_size=batch_size,
                         vocab_threshold=vocab_threshold,
                         vocab_from_file=vocab_from_file)

# The size of the vocabulary.
vocab_size = len(data_loader.dataset.vocab)

# Initialize the encoder and decoder. 
encoder = EncoderCNN(embed_size)
decoder = DecoderRNN(embed_size, hidden_size, vocab_size)

# Move models to GPU if CUDA is available. 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder.to(device)
decoder.to(device)

# Define the loss function. 
criterion = nn.CrossEntropyLoss().cuda() if torch.cuda.is_available() else nn.CrossEntropyLoss()

# TODO #3: Specify the learnable parameters of the model.
params = list(decoder.parameters()) + list(encoder.embed.parameters()) 

# TODO #4: Define the optimizer.
optimizer = torch.optim.Adam(params=params, lr = 0.001)

# Set the total number of training steps per epoch.
total_step = math.ceil(len(data_loader.dataset.caption_lengths) / data_loader.batch_sampler.batch_size)
````

````python
import torch.utils.data as data
import numpy as np
import os
import requests
import time
from workspace_utils import *

# Open the training log file.
f = open(log_file, 'w')

old_time = time.time()
response = requests.request("GET", 
                            "http://metadata.google.internal/computeMetadata/v1/instance/attributes/keep_alive_token", 
                            headers={"Metadata-Flavor":"Google"})

with active_session():
    for epoch in range(1, num_epochs+1):

        for i_step in range(1, total_step+1):

            if time.time() - old_time > 60:
                old_time = time.time()
                requests.request("POST", 
                                 "https://nebula.udacity.com/api/v1/remote/keep-alive", 
                                 headers={'Authorization': "STAR " + response.text})

            # Randomly sample a caption length, and sample indices with that length.
            indices = data_loader.dataset.get_train_indices()
            # Create and assign a batch sampler to retrieve a batch with the sampled indices.
            new_sampler = data.sampler.SubsetRandomSampler(indices=indices)
            data_loader.batch_sampler.sampler = new_sampler

            # Obtain the batch.
            images, captions = next(iter(data_loader))

            # Move batch of images and captions to GPU if CUDA is available.
            images = images.to(device)
            captions = captions.to(device)

            # Zero the gradients.
            decoder.zero_grad()
            encoder.zero_grad()

            # Pass the inputs through the CNN-RNN model.
            features = encoder(images)
            outputs = decoder(features, captions)

            # Calculate the batch loss.
            loss = criterion(outputs.view(-1, vocab_size), captions.view(-1))

            # Backward pass.
            loss.backward(retain_graph=True)

            # Update the parameters in the optimizer.
            optimizer.step()
            optimizer.zero_grad()

            # Get training statistics.
            stats = 'Epoch [%d/%d], Step [%d/%d], Loss: %.4f, Perplexity: %5.4f' % (epoch, num_epochs, i_step, total_step, loss.item(), np.exp(loss.item()))

            # Print training statistics (on same line).
            print('\r' + stats, end="")
            sys.stdout.flush()

            # Print training statistics to file.
            f.write(stats + '\n')
            f.flush()

            # Print training statistics (on different line).
            if i_step % print_every == 0:
                print('\r' + stats)

            # Save the weights.
            if i_step % print_every == 0:
                torch.save(decoder.state_dict(), os.path.join('./models', 'decoder-%d.pkl' % epoch))
                torch.save(encoder.state_dict(), os.path.join('./models', 'encoder-%d.pkl' % epoch))

# Close the training log file.
f.close()
````
